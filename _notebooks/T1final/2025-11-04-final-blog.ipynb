# **CSP Final Blog / Semester Reflection**

This semester in CSP has been one of the biggest transformations in my academic journey. At the start of the year, my understanding of computer science was extremely limited. I didn’t know how to use a terminal, I had no idea how repositories were structured, and GitHub seemed like an advanced tool that only experienced programmers could navigate. Over the months, I learned not only how to control the terminal confidently, but also how to understand the logic behind commands and how they interact with files and code.  

Through practice and repetition, I became comfortable with working inside repositories, switching between directories, managing files, and collaborating through GitHub. I also became much more independent when solving errors. At the beginning of the year, any error message felt overwhelming — but now, I can slow down, think through what the error means, and fix it on my own or with minimal help. This shift has made me much more confident in my coding abilities.

A huge part of my growth came from collaboration. I made new friends in CSP, and working with people helped me understand concepts that would have been harder alone. Debugging together taught me real teamwork and strengthened my communication skills. Whenever we faced a confusing bug, instead of panicking, we learned to break the problem down and approach it logically. By the end of the semester, I felt a much deeper understanding of what it means to “think like a programmer.”

---

## **Growth Across the Sprints**

### **Tools Unit**
During the Tools unit, I learned how to navigate efficiently through different project directories using the terminal. Commands like `ls`, `cd`, `mkdir`, `mv`, `rm`, `git clone`, `git status`, `git add`, and `git push` went from being confusing words to tools I used every day. I learned not only *what* the commands do, but *why* they are essential. This unit taught me how developers manage their workflow professionally and gave me the confidence to work in a real development environment.

### **Fundamentals of Python & JavaScript**
In the Fundamentals sprint, I developed a strong base in both Python and JavaScript. I became comfortable using loops, variables, lists/arrays, conditional statements, and writing functions that return values. I also gained experience interpreting pseudocode, reading error messages, and translating logic from one language into another. Understanding control flow and data handling helped me prepare for more complex programming problems, and I realized how important it is to understand the “why” behind each line of code.

### **West Coast Travel / Digital Famine Project**
This project deepened my understanding of GitHub submodules — a concept that at first felt extremely difficult. I learned how submodules allow you to connect multiple repositories inside one main repository, and how that structure is used in real-world projects. On the web design side, I experimented with layout designs, CSS styling, color palettes, typography, and content organization. I gained a better sense of what makes a webpage visually appealing and easy for users to understand. This sprint taught me creativity, problem-solving, and communication through design.

## **Night at the Museum Reflection**

Night at the Museum was a highlight of the semester. When people visited our table, they commented on how clean and organized our page looked. Parents appreciated how easy it was to read and navigate, and guests complimented the structure and visuals of our website. It was rewarding to watch people interact with something we created from scratch. Presenting the project helped me practice explaining technical work in a simple and confident way. I realized that communication and presentation are major parts of computer science — not just writing the code but showing the purpose behind it.
