# **CSP Final Blog / Semester Reflection**

This semester in CSP has been one of the biggest academic transformations I’ve had so far. At the beginning of the year, I barely knew anything about computer science. I didn’t understand how repositories worked, I didn’t know how to use terminal commands, and GitHub felt like a completely foreign tool. Now, I can move through directories quickly, clone and push repositories, and use a whole set of terminal commands that used to intimidate me. I also learned how version control works behind the scenes — how branches, commits, pull requests, and merges all fit together.  

Along the way, I met new people in class and collaborated with them on coding tasks, debugging sessions, and project building. Working with others helped me understand concepts faster and made the entire learning process more enjoyable. Looking back, I can see how much more confident and comfortable I’ve become in problem-solving situations, especially when something breaks and I need to figure out why.

# **MCQ Review (55/66)**

---

## **Q1 — Swapping Values**
I misunderstood the correct order of operations in a swap. A proper swap saves the original value, overwrites the first variable, then restores the saved value.  
I chose a step that overwrote the wrong variable.

✅ **Remember:** swapping always follows  
**temp ← first → first ← second → second ← temp**

---

## **Q12 — Social Media Year 12 Projection**
I estimated the growth incorrectly. The total number of users was increasing in a roughly linear pattern, about **0.6 per year**, not randomly. This meant that projecting two more years should land around **31.2**, not 31.8.

✅ **Remember:** when predicting trends, use **average change**, not individual spikes.

---

## **Q15 — Internet Communication**
I incorrectly thought devices must have the same operating system.  
But the Internet is built on **shared protocols**, like TCP/IP, that work across all systems.

✅ **Remember:** every device on the Internet has an **IP address**, not a specific OS.

---

## **Q16 — Licensed Music Download Consequences**
I overestimated how “unlikely" targeted ads are — they are actually *very* likely.  
The least likely consequence was users seeing reviews from people nearby.

✅ **Remember:** ads follow your behavior constantly; location-based review targeting rarely happens.

---

## **Q26 — Robot Movement (Missing Code)**
I picked an option that didn’t correctly handle turning or checking forward movement. The correct program avoided black squares by checking forward and adjusting the robot’s rotation only when needed.

✅ **Remember:** robot logic =  
**check → move → rotate** (only when blocked)

---

## **Q31 — Loop Algorithms With Robots**
I assumed Program I reached the goal, but tracing carefully shows that **neither program** actually reaches the gray square.

✅ **Remember:** manually walk through robot programs step-by-step — don’t assume loops succeed.

---

## **Q43 — Runtime of Algorithm**
I misread the purpose of the question. The runtime was growing quadratically (n²), which still counts as “reasonable time.”

✅ **Remember:**  
- Polynomial → reasonable  
- Exponential → unreasonable  
- Approximation has nothing to do with runtime here

---

## **Q45 — NAND Logic Gate**
I selected the wrong Boolean expression.  
The correct one is:

✅ **`NOT (P AND Q)`**

✅ **Remember:** NAND is simply the opposite of AND.

---

## **Q46 — Undecidable Problems**
I chose an answer saying the algorithm could exist in a low-level language. This is incorrect. For undecidable problems, **no algorithm can solve it for all inputs**, in any language.

✅ **Remember:** undecidable = impossible algorithmically.

---

## **Q53 — Machine Learning Program Purpose**
I picked a choice about approximate solutions, but machine learning is actually about analyzing data and finding patterns.

✅ **Remember:**  
ML = **patterns, recognition, prediction**, not solving undecidable problems.
